```python
import pandas as pd
```

```python
def read_excel_columns(file_path, sheet,columns):
   
    df = pd.read_excel(file_path, sheet_name=sheet, usecols=columns,  engine='openpyxl')

    return df
```

```python
# read NRCELL sheet
NRCELL_COLUMNS = ["MO", "name", "lcrId",'physCellId','freqBandIndicatorNR','nrarfcn','chBw','configuredEpsTac',
                  'gscnOrSsPbchArfcn','mMimoAntArrayMode','ssbScs','prachRootSequenceIndex','frRefList','modDNList','ulDlDataSlotRatio']
file = 'Export_MultiTable_5G.xlsx'
nrcell_df = read_excel_columns(file, 'NRCELL', NRCELL_COLUMNS)
nrcell_df.head()
```

```python
# read INVUNIT
INVUNIT_COLUMNS = ["configDN", "inventoryUnitType"]

INVUNIT_DF = read_excel_columns(file, 'INVUNIT', INVUNIT_COLUMNS)
INVUNIT_DF.dropna(inplace=True)
INVUNIT_DF.head()
```

```python
df_final = pd.merge(nrcell_df, INVUNIT_DF, left_on='modDNList', right_on='configDN', how='left')
df_final.drop(columns=['configDN'], inplace=True)
df_final.rename(columns={'inventoryUnitType':'BBU'}, inplace=True)
df_final.head()
```

```python
df_final = pd.merge(df_final, INVUNIT_DF, left_on='frRefList', right_on='configDN', how='left')
df_final.drop(columns=['configDN'], inplace=True)
df_final.rename(columns={'inventoryUnitType':'RRU'}, inplace=True)
df_final.head()
```

```python
df_final['RMOD'] = df_final['frRefList'].astype(str).str.replace(r'^.*/', '', regex=True)  
df_final.head()
```

```python
df_final.drop(columns=['frRefList','modDNList'],inplace= True)
df_final.head()
```

```python
df_final['mMimoAntArrayMode'].unique()
```

```python
replacement_dict = {
    'Full 32TRX Array (2x8x2)': '32T32R',
    'Full 64TRX Array (4x8x2)': '64T64R',}
df_final['mMimoAntArrayMode'] = df_final['mMimoAntArrayMode'].replace(replacement_dict)
df_final.head()
```

```python
df_final['freqBandIndicatorNR']='N77'
df_final.head()
```

```python
replacement_dict = {
    'ABIP AirScale Capacity': 'ABIP'}
df_final['BBU'] = df_final['BBU'].replace(replacement_dict)
df_final.head()
```

```python
df_final['gNodeB ID'] = df_final['MO'].astype(str).str.replace(r'/NRCEL.*', '', regex=True).str.replace(r'^.*-', '', regex=True)
df_final.head()
```

```python
df_final['Base Station Name'] = df_final['name'].astype(str).str.slice(0,6)
df_final.head()
```

```python
re_name_dict = {
    'name':'Cell Name','configuredEpsTac':'TAC','physCellId':'PCI',
    'nrarfcn':'DL NARFCN','chBw':'DL Bandwidth','gscnOrSsPbchArfcn':'SSB FP',
    'mMimoAntArrayMode':'TxRx Mode','ssbScs':'SCS (KHz)','freqBandIndicatorNR':'Band','prachRootSequenceIndex':'RSI',
    'ulDlDataSlotRatio':'Slot Assignment',"lcrId":'Cell ID'}
df_final.rename(columns=re_name_dict, inplace=True)
df_final.head()
```

```python
replacement_dict = {
    '1/4': '8_2_DDDDDDDSUU'}
df_final['Slot Assignment'] = df_final['Slot Assignment'].replace(replacement_dict)
df_final.head()
```

```python
df_final = df_final[['Base Station Name','Cell ID','Cell Name','PCI','Band','DL NARFCN','DL Bandwidth','SCS (KHz)',
                     'Slot Assignment','TAC','SSB FP','RSI','TxRx Mode','RMOD','BBU','RRU','gNodeB ID']]
df_final.head()
```

```python
df_final['Vendor'] = 'Nokia'
df_final.head()
```

```python
df_final.to_excel('5G_NSN_inventory_output.xlsx', index=False)
```
