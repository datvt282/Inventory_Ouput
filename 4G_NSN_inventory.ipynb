# file Export_MultiTable bao gồm các sheet
 1.LNCEL_FDD
 2.LNCEL
 3.INVUNIT
 4.CHANNEL

```python
import pandas as pd
```

```python
def read_excel_columns(file_path, sheet,columns):
   
    df = pd.read_excel(file_path, sheet_name=sheet, usecols=columns,  engine='openpyxl')

    return df
```

```python
# read LNCEL_FDD
LNCEL_FDD_COLUMNS = ["MO", "earfcnDL", "prachCS", "rootSeqIndex", "dlChBw", "dlMimoMode"]
file = 'Export_MultiTable.xlsx'
LNCEL_FDD_DF = read_excel_columns(file, 'LNCEL_FDD', LNCEL_FDD_COLUMNS)
LNCEL_FDD_DF.head()
```

```python
LNCEL_FDD_DF['MO'] = LNCEL_FDD_DF['MO'].astype(str).str.replace('/LNCEL_FDD-0', '', regex=False)
```

```python
LNCEL_FDD_DF.head()
```

```python
# Đổi tên các giá trị trong cột dlMimoMode
replace_dict = {
    'Closed Loop Mimo': '2T2R',
    'Closed Loop MIMO (4x4)': '4T4R',
    'SingleTX': '1T1R'
}
LNCEL_FDD_DF['dlMimoMode'] = LNCEL_FDD_DF['dlMimoMode'].replace(replace_dict)
LNCEL_FDD_DF.head()
```

```python
# read LNCEL
LNCEL_COLUMNS = ["MO", "name", "phyCellId",'tac']

LNCEL_DF = read_excel_columns(file, 'LNCEL', LNCEL_COLUMNS)
LNCEL_DF.head()
```

```python
df_final = pd.merge(LNCEL_DF, LNCEL_FDD_DF, on='MO', how='inner')
df_final.head()
```

```python
# read INVUNIT
INVUNIT_COLUMNS = ["MO", "configDN", "inventoryUnitType"]

INVUNIT_DF = read_excel_columns(file, 'INVUNIT', INVUNIT_COLUMNS)
INVUNIT_DF.head()
```

```python
INVUNIT_DF['MO'] = INVUNIT_DF['MO'].astype(str).str.replace(r'/HW.*$', '', regex=True)   
INVUNIT_DF.head()
```

```python
# Đổi tên các giá trị trong cột inventoryUnitType
replace_dict = {
    'Flexi Baseband Sub-Module FBBC': 'FBBC',
    'Flexi System Module Outdoor FSMF': 'FSMF',
    'BB Extension Outdoor Sub-Module FBBA': 'FBBA'
}
INVUNIT_DF['inventoryUnitType'] = INVUNIT_DF['inventoryUnitType'].replace(replace_dict)
INVUNIT_DF.head()
```

```python
# Chỉ giữ lại các giá trị inventoryUnitType mong muốn trong INVUNIT_DF và tách thành 2 DataFrame riêng cho BB và RF
BB_unit_types = ['FBBC', 'FSMF','FBBA']
RF_unit_types = ['FHEL', 'AHEB', 'FRGY', 'FXED', 'FRGU', 'FRGX', 'FRHC', 'FRGT',  'FRGP']
BB_INVUNIT_DF = INVUNIT_DF[INVUNIT_DF['inventoryUnitType'].isin(BB_unit_types)]
BB_INVUNIT_DF.head()
  
```

```python
RF_INVUNIT_DF = INVUNIT_DF[INVUNIT_DF['inventoryUnitType'].isin(RF_unit_types)]
RF_INVUNIT_DF.head()  
```

```python
BB_INVUNIT_DF['BB_count'] = BB_INVUNIT_DF.groupby(['MO','inventoryUnitType'])['inventoryUnitType'].transform('count')
BB_INVUNIT_DF.head()
```

```python
BB_INVUNIT_DF.to_excel('BB_INVUNIT_DF_4G_NSN.xlsx', index=False)
```

```python
BB_INVUNIT_DF['BB_map'] = BB_INVUNIT_DF['BB_count'].astype(str) + '_' + BB_INVUNIT_DF['inventoryUnitType']
BB_INVUNIT_DF.head()
```

```python
# Sắp xếp BB_INVUNIT_DF theo inventoryUnitType giảm dần
BB_INVUNIT_DF = BB_INVUNIT_DF.sort_values(by=['inventoryUnitType','MO'], ascending=False)
BB_INVUNIT_DF.head()
```

```python
BB_INVUNIT_DF = BB_INVUNIT_DF[['MO', 'BB_map']].drop_duplicates()
BB_INVUNIT_DF.head()
```

```python
BB_INVUNIT_DF['BB_configDN'] = BB_INVUNIT_DF.groupby(['MO'])['BB_map'].transform(lambda x: ' + '.join(x))
BB_INVUNIT_DF.head()
```

```python
RF_INVUNIT_DF.head()
```

```python
RF_INVUNIT_DF.to_excel('RF_INVUNIT_DF_4G_NSN.xlsx', index=False)
```

```python
RF_INVUNIT_DF['RMOD'] = RF_INVUNIT_DF['configDN'].astype(str).str.replace(r'^.*/', '', regex=True)   
RF_INVUNIT_DF.head()   
```

```python
RF_INVUNIT_DF.drop(columns=['MO'], inplace=True)
RF_INVUNIT_DF.head()
```

```python
# read CHANNEL
CHANNEL_COLUMNS = ["MO", "antlDN", "direction"]

CHANNEL_DF = read_excel_columns(file, 'CHANNEL', CHANNEL_COLUMNS)
CHANNEL_DF.head()
```

```python
CHANNEL_DF = CHANNEL_DF[CHANNEL_DF['direction'].isin(['TX'])]
CHANNEL_DF.head()
```

```python
CHANNEL_DF['antlDN'] = CHANNEL_DF['antlDN'].astype(str).str.replace(r'/AN.*$', '', regex=True)   
CHANNEL_DF.head()
```

```python
df_final_2  = pd.merge(CHANNEL_DF,RF_INVUNIT_DF, left_on='antlDN', right_on='configDN', how='inner')
df_final_2.head()
```

```python
df_final_2.drop(columns=['antlDN','direction','configDN'], inplace=True)
df_final_2.head()
```

```python
df_final_2['MO'] = df_final_2['MO'].astype(str).str.replace('/MNL-1/MNLENT-1/CELLMAPPING-1', '', regex=False)
df_final_2.head()
```

```python
df_final_2['MO'] = df_final_2['MO'].astype(str).str.replace(r'/CHAN.*$', '', regex=True)
df_final_2.drop_duplicates(inplace=True)
df_final_2.head()
```

```python
df_final_2['MO'] = df_final_2['MO'].astype(str).str.replace('LCELL', 'LNCEL', regex=False)
df_final_2.head()
```

```python
df_final.head()
```

```python
df_final['MO'] = df_final['MO'].astype(str).str.replace(r'LNBTS.*/', '', regex=True)
df_final.head()
```

```python
df_final_3 = pd.merge(df_final_2, df_final, on='MO', how='left')
df_final_3.head()
```

```python
df_final_3["eNodeB_id"] = df_final_3['MO'].astype(str).str.replace(r'/LNCEL-(\d+)', '', regex=True)
df_final_3.head()   
```

```python
# Tạo cột Band trong df_final_3 dựa trên giá trị earfcnDL
band_map = {1501: 3, 1874: 3, 50: 1, 3150: 7, 1300:3}
df_final_3['Band'] = df_final_3['earfcnDL'].map(band_map)
df_final_3.head()
```

```python
df_final_3['RF type Count'] = df_final_3.groupby(['eNodeB_id','RMOD','inventoryUnitType','Band'])['inventoryUnitType'].transform('count')
df_final_3.head()
```

```python
import numpy as np
df_final_3['RF_Type_Count_Inv'] = np.where(
    df_final_3['RF type Count'] == 1, "1",
    np.where(df_final_3['RF type Count'] == 2, "0.5",
    np.where(df_final_3['RF type Count'] == 4, "0.25",
    np.where(df_final_3['RF type Count'] == 6, "1/6",       
    np.where(df_final_3['RF type Count'] == 3, "1/3", 1/df_final_3['RF type Count'])))))
df_final_3.head(2)
```

```python
df_final_3['Config_RRU_per_Cell'] = df_final_3['RF_Type_Count_Inv'].astype(str) + '_' + df_final_3['inventoryUnitType'].astype(str)
df_final_3.head() 
```

```python
df_final_3['Config_RRU'] = (df_final_3.groupby(['MO','name'])['Config_RRU_per_Cell'].transform(lambda x: "+".join(map(str, x))))
```

```python
df_final_3 = df_final_3.sort_values(by='RMOD',ascending=True)
df_final_3['Subrack'] = (df_final_3.groupby(['MO','name'])['RMOD'].transform(lambda x: "+".join(map(str, x))))
df_final_3.head()
```

```python
df_final_3['localcell ID'] = df_final_3['MO'].replace(r'^.*-', '', regex=True)
```

```python

df_final_3['eNodeB_id'] = df_final_3['eNodeB_id'].replace(r'^.*-', '', regex=True)
df_final_3['MO'] = df_final_3['MO'].replace(r'/LN.*', '', regex=True)
df_final_3.head()
```

```python
df_final_3['Vendor']= 'Nokia'
```

```python
df_final_4 = pd.merge(df_final_3, BB_INVUNIT_DF[['MO', 'BB_configDN']], on='MO', how='left')
df_final_4.head()
```

```python
rename_columns = {'name':'Cell Name', 'phyCellId':'PCI', 'earfcnDL':'EARFCN',
                  'dlChBw':'Bandwidth', 'dlMimoMode':'MIMO','rootSeqIndex':'RSI', 'tac':'TAC',
                  'BB_configDN':'Config_BBU','prachCS':'Cell Radius','localcell ID':'Local Cell ID','MO':'Base Station Name',
                  'eNodeB_id':'eNodeB ID'
                  } 
df_final_4.rename(columns=rename_columns, inplace=True)
df_final_4.head()
```

```python
df_final_Nokia = df_final_4[['Base Station Name', 'Local Cell ID', 'Cell Name', 'Band', 'EARFCN',
       'Bandwidth', 'MIMO','eNodeB ID', 'TAC', 'PCI', 'RSI', 'Cell Radius', 'Vendor', 'Subrack', 'Config_BBU', 'Config_RRU']]
df_final_Nokia.head()
```

```python
df_final_Nokia['Local Cell ID']= df_final_Nokia['Local Cell ID'].astype(int)
df_final_Nokia['eNodeB ID']= df_final_Nokia['eNodeB ID'].astype(int)
df_final_Nokia.drop_duplicates(inplace=True)
df_final_Nokia.head()
```

```python
df_final_Nokia.to_excel('4G_NSN_Inventory_Output.xlsx', index=False)
```
