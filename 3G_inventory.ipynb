```python
def read_excel_columns(file_path, columns, header_row=0,):
   
    df = pd.read_excel(file_path, header=header_row,engine='openpyxl')
    df = df[[col for col in columns if col in df.columns]]

    return df
```

```python
def read_file_skip_headers(folder,sheet_name,columns=None):
    files = sorted(glob.glob(folder + "/*.xlsx"))
    first_df = pd.read_excel(files[0], sheet_name=sheet_name)
    dfs = [first_df]

    for file in files[1:]:
        d = pd.read_excel(file, sheet_name=sheet_name,header=5)
        d.columns = first_df.columns
        dfs.append(d)
    df_result = pd.concat(dfs, ignore_index=True)
    if columns is not None:
        df_result = df_result[[col for col in columns if col in df_result.columns]]
    return df_result
```
